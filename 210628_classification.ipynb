### Laplace smoothing

x: 샘플
$$\theta_i = \frac{x_i+\alpha}{N+\alpha d}$$


$$i \in {red, green, blue}$$
$$\theta_i = \frac{x_i}{N}$$

$$x_{red} = 6$$
$$x_{green} = 4$$

$$\theta_{red} = \frac{6}{10}$$
$$\theta_{green} = \frac{4}{10}$$
$$\theta_{blue} = \frac{0}{10} = 0$$

$$\alpha = 0.1, d = 3$$

$$\theta_i = \frac{x_i+\alpha}{N+\alpha d}$$


$$\theta_{red} = \frac{6 + \alpha}{10 + 3\alpha} = \frac{6 + 0.1}{10 + 0.3} \sim 0.5922$$
$$\theta_{green} = \frac{4 + \alpha}{10 + 3\alpha} = \frac{4 + 0.1}{10 + 0.3} \sim 0.3981$$
$$\theta_{blue} = \frac{0 + \alpha}{10 + 3\alpha} = \frac{0 + 0.1}{10 + 0.3} \sim 0.0097 $$

$$ M_A : \theta_{red} = 0.7, \theta_{green} = 0.2, \theta_{blue} = 0.1 $$
$$ M_B : \theta_{red} = 0.3, \theta_{green} = 0.4, \theta_{blue} = 0.3 $$

$${red: 4, blue: 3, green: 3}$$
$$P(M_A) = 0.7, P(M_B) = 0.3$$

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

x: 뽑혀진 공 10개 샘플(표집값)
$$k \in {A, B}$$

---

$$P(M_k|x) = \frac{P(x|M_k) * P(M_k)}{P(x)}$$

$$P(M_A|x) \propto P(M_A) * P(x|M_A) = 0.7 * (0.7^4 * 0.1^3 * 0.2^3) = 1.345 * 10^{-6}$$
$$P(M_A|x) \propto 1.345 * 10^{-6}$$

$$P(M_B|x) \propto P(M_B) * P(x|M_B) = 0.3 * (0.3^4 * 0.4^3 * 0.3^3) = 4.199 * 10^{-6}$$
$$P(M_B|x) \propto 4.199 * 10^{-6}$$

```python
import re

input_train_sentences = [
                         "John likes to watch movies. Sam likes movies too.",
                         "Machine learning is a method of watch data analysis that automates analytical model building."
                         ]

input_train_sentences_prob = [0.5, 0.5]
test_sentence = "John also likes to watch football games."
alpha = 0.1

def make_bow(t_sentence):
  word_dict = {}
  word_cnt = []

  t_sentence = t_sentence.lower()
  t_sentence = re.sub(r'[^a-z]+', ' ', t_sentence)

  for word in t_sentence.split():
    if word not in word_dict:
      word_dict[word] = len(word_cnt)
      word_cnt.append(0)
    
    word_cnt[word_dict[word]] += 1 

  return word_dict, word_cnts

def calc_sentence_prob(word_model_dict, word_model_cnt, test_model_dict, test_model_cnt, alpha):
  sentence_prob = 1.0

  for word in test_model_dict:
    test_word_freq_cnt = test_model_cnt[test_model_dict[word]]

    train_word_freq_cnt = 0
    if word in word_model_dict:
      train_word_freq_cnt = word_model_cnt[word_model_dict[word]]

    N = sum(word_model_cnt)
    d = len(word_model_cnt)

    prob = ((train_word_freq_cnt + alpha) / (N + alpha * d)) ** test_word_freq_cnt
    sentence_prob *= prob

  return sentence_prob

# print(input_train_sentences)

sentence_prob_list = []

for sentence in input_train_sentences:
  boword_model_dict, boword_model_cnt = make_bow(sentence)
  test_model_dict, test_model_cnt = make_bow(test_sentence)
  
  sentence_prob_list.append(calc_sentence_prob(boword_model_dict, boword_model_cnt, test_model_dict, test_model_cnt, alpha))

total_prob = sum(sentence_prob_list)
for idx, prob in enumerate(sentence_prob_list):
  sentence_prob_list[idx] = prob / total_prob

print(sentence_prob_list)
```
